```python
import pandas as pd
titanic_df = pd.read_csv(r'C:/titanic dataset.csv')
titanic_df.head()
```

```python
# Checking missing values
titanic_df.isnull().sum()
```

```python
# Fill missing 'Age' values with the median age
titanic_df.loc[:, 'Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())

# Fill missing 'Embarked' values with the most common port
titanic_df.loc[:, 'Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])

# Drop the 'Cabin' column if it exists
if 'Cabin' in titanic_df.columns:
    titanic_df.drop(columns=['Cabin'], inplace=True)
```
